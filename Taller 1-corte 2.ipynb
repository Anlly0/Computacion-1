```python
import random
import matplotlib.pyplot as plt

def simulador_asientos():
    asientos = [False] * 100
    asientos[random.randint(0, 99)] = True
    for i in range(1, 99):
        if asientos[i]:
            while True:
                asiento_aleatorio = random.randint(0, 99)
                if not asientos[asiento_aleatorio]:
                    asientos[asiento_aleatorio] = True
                    break
        else:
            asientos[i] = True
    return asientos.index(False) + 1  

resultados = [simulador_asientos() for _ in range(1_000_000)]

plt.hist(resultados, bins=100, edgecolor='black', alpha=0.7)
plt.xlabel('Número de asiento ocupado por el pasajero 100')
plt.ylabel('Frecuencia')
plt.title('Distribución de asientos del pasajero 100 en 1 millón de simulaciones')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
