```python
class Producto:
    def __init__(self, nombre, precio, categoria):
        self.nombre = nombre
        self.precio = precio
        self.categoria = categoria

    def __str__(self):
        return f"{self.nombre} - ${self.precio}"

class Tienda:
    def __init__(self, nombre):
        self.nombre = nombre
        self.productos = []
        print(f'__{self.nombre}__')
        self.mostrar_productos()

    def agregar_productos(self, producto):
        self.productos.append(producto)
        print(f'\nAgregar productos: {producto}')
        self.mostrar_productos()

    def mostrar_productos(self):
        print('\nProductos de la tienda:')
        if not self.productos:
            print("No hay productos a√∫n.")
        else:
            for p in self.productos:
                print(p)
    def buscar_producto(self, nombre):
        for producto in self.productos:
            if producto.nombre.lower() == nombre.lower():
                return producto
        return None
    def producto_mas_caro(self):
        if not self.productos:
            return None
        return max(self.productos, key=lambda p: p.precio)

    def producto_mas_barato(self):
        if not self.productos:
            return None
        return min(self.productos, key=lambda p: p.precio)

    def ordenar_por_precio(self, descendente=False):
        return sorted(self.productos, key=lambda p: p.precio, reverse=descendente)

pan = Producto("Pan", 500, "Comidas")
jabon = Producto("Jab√≥n", 1200, "Aseo")
galletas = Producto("galletas", 7000, "comidas")
leche = Producto("leche", 5000, "comidas")

print("\n üö©Mostrar productos de la tienda A:")
tiendaA = Tienda("Supermercado")
tiendaA.agregar_productos(pan)
tiendaA.agregar_productos(jabon)


tiendaB = Tienda("lola")
print("\n üö©Mostrar productos de la tienda B:")
tiendaB.agregar_productos(leche)
tiendaB.agregar_productos(galletas)




buscado = tienda.buscar_producto("azucar")
if buscado:
    print(f"\n‚úÖ Producto encontrado: {buscado}")
else:
    print("\n‚ùå Producto no encontrado")


caro = tienda.producto_mas_caro()
if caro:
    print(f"\nüí∞ Producto m√°s caro: {caro}")


barato = tienda.producto_mas_barato()
if barato:
    print(f"üí∏ Producto m√°s barato: {barato}")


print("\nüìä Productos ordenados por precio (menor a mayor):")
for prod in tienda.ordenar_por_precio():
    print(prod)


print("\nüìä Productos ordenados por precio (mayor a menor):")
for prod in tienda.ordenar_por_precio(descendente=True):
    print(prod)
```
